```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.arange(0,10)
y = x**2
```

```python
# plot x and y data
plt.plot(x, y)
```

```python
# plot a subset of the data
plt.plot(x, y)
plt.xlim(0,5)
plt.ylim(0,40)
```

```python
# add labels
plt.plot(x, y)
plt.title('Sample Title')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```
